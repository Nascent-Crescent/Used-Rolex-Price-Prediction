```python
import pandas as pd
df = pd.read_csv('cleaned_rolex.csv')
```

```python
import matplotlib.pyplot as plt


boxplot = df.boxplot(column='price', by='model', figsize=(25,12))

plt.title('Box Plot of Price Grouped by Models', fontsize=16)
plt.suptitle("")  # Remove the default group label
plt.xlabel('Model', fontsize=14)
plt.ylabel('Price', fontsize=14)

# Manually set the x-axis tick positions and labels for more spread-out ticks
x_tick_positions = range(1, len(df['model'].unique()) + 1)  # Adjust tick positions
x_tick_labels = df['model'].unique()  # Use unique values from the 'model' column
plt.xticks(x_tick_positions, x_tick_labels, rotation=45, ha='right', fontsize=12)

plt.subplots_adjust(bottom=0.2)  # Adjust bottom margin for rotated labels

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

We can see that the price has a lot of outliers. This will affect our modeling process later on. One way to deal with this is to remove entries with price higher than 99 percentile of the given model.

```python
import numpy as np

def filter_rows(group):
    q = group['price'].quantile(.99)
    return group[group['price'] <= q]

df_filtered = df.groupby('model').apply(filter_rows).reset_index(drop=True)
```

```python
boxplot = df_filtered.boxplot(column='price', by='model', figsize=(25,12))

plt.title('Box Plot of Price Grouped by Models', fontsize=16)
plt.suptitle("")  # Remove the default group label
plt.xlabel('Model', fontsize=14)
plt.ylabel('Price', fontsize=14)

# Manually set the x-axis tick positions and labels for more spread-out ticks
x_tick_positions = range(1, len(df['model'].unique()) + 1)  # Adjust tick positions
x_tick_labels = df['model'].unique()  # Use unique values from the 'model' column
plt.xticks(x_tick_positions, x_tick_labels, rotation=45, ha='right', fontsize=12)

plt.subplots_adjust(bottom=0.2)  # Adjust bottom margin for rotated labels

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

From the boxplot, we see that there are much fewer outliers in our data. Mission accomplished.

Now we one hot code all categorical data.

```python
import sklearn
categorical_cols = ['model','movement','case material','condition','scope of delivery','country']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_filtered[categorical_cols]  = df_filtered[categorical_cols].apply(lambda col: le.fit_transform(col))

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()

array_hot_encoded  = ohe.fit_transform(df_filtered[categorical_cols])
df_hot_encoded = pd.DataFrame(array_hot_encoded)

df_hot_encoded = pd.concat([df_hot_encoded,df_filtered.drop(columns= categorical_cols)], axis = 1)
```
